```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('Area (in Sq meters)')
plt.ylabel('Price (in US $)')
plt.scatter(df.area, df.price,color='red', marker= '+')
```

```python
# Our data fit for linear Regression model So let us create Linear Regression model
# Create an object for linear Regression
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
135.78767123*3300+180616.43835616432
```

```python
x = pd.DataFrame({'area':[3300,4000]})
reg.predict(x)
```

```python
p= pd.read_csv('areas.csv')
```

```python
d = reg.predict(p)
```

```python
p['price'] = d
```

```python
p.to_csv("Predictions.csv")
```

```python
# now showing how my scattered data with Linear Prediction line
%matplotlib inline
plt.xlabel('Area (in Sq meters)')
plt.ylabel('Price (in US $)')
plt.scatter(df.area, df.price,color='red', marker= '+')
plt.plot(df.area,reg.predict(df[['area']]),color= 'blue')
```
